# **Agrupar i Agregar**

---



```python
ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
soldprice_by_age = ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age)
# Ordena el valors pel preu de venda.
```

```python
soldprice_by_age_role = ipl_auction_df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age_role)
```

# **Unió de Dataframes**

---



```python
soldprice_comparison = soldprice_by_age_role.merge(
    soldprice_by_age,
    on='AGE',
    how='outer'
)
soldprice_comparison
```

```python
soldprice_comparison.rename(
    columns={
        'SOLD PRICE_x': 'SOLD_PRICE_AGE_ROLE',
        'SOLD PRICE_y': 'SOLD_PRICE_AGE'
    },
    inplace=True
)

soldprice_comparison.head(5)
```

```python
soldprice_comparison['change'] = soldprice_comparison.apply(
    lambda rec: (rec.SOLD_PRICE_AGE_ROLE - rec.SOLD_PRICE_AGE) / rec.SOLD_PRICE_AGE,
    axis=1
)
soldprice_comparison
```
