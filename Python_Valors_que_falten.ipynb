# **Valors que falten**

---
Al m√≥n real, els conjunts de dades no estan nets i poden tenir valors que

falten. Hem de saber trobar i tractar amb aquests valors que falten.

El dataset pot ser descarregat de https://archive.ics.uci.edu/ml/datasets/auto+mpg



```python
import pandas as pd
autos = pd.read_csv('/content/auto-mpg.data', sep = '\s+', header = None)
autos.head(5)
```

```python
autos.columns = [
    'mpg', 'cylinders', 'displacement',
    'horsepower', 'weight', 'acceleration',
    'year', 'origin', 'name'
]
autos.head(5)
```

```python
autos.info()

# Aqu√≠ s'ha dedu√Øt la pot√®ncia de la columna com a objecte, mentre que s'hauria d'haver dedu√Øt com a float64.
# Aix√≤ pot ser perqu√® algunes de les files contenen valors no num√®rics a la columna de pot√®ncia.
# Una opci√≥ per fer-ho √©s for√ßar la conversi√≥ de la columna de cavalls de pot√®ncia en num√®rica,
# que hauria de convertir els valors no num√®rics en valors nuls.
```

```python
autos['horsepower'] = pd.to_numeric(autos['horsepower'], errors='coerce')
autos.info()
```

```python
autos[autos.horsepower.isnull()]
```

```python
autos = autos.dropna(subset = ['horsepower'])
autos[autos.horsepower.isnull()]
```
