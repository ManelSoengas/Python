# **Valors que falten**

---
Al món real, els conjunts de dades no estan nets i poden tenir valors que

falten. Hem de saber trobar i tractar amb aquests valors que falten.

El dataset pot ser descarregat de https://archive.ics.uci.edu/ml/datasets/auto+mpg



```python
import pandas as pd
autos = pd.read_csv('/content/auto-mpg.data', sep = '\s+', header = None)
autos.head(5)
```

```python
autos.columns = [
    'mpg', 'cylinders', 'displacement',
    'horsepower', 'weight', 'acceleration',
    'year', 'origin', 'name'
]
autos.head(5)
```

```python
autos.info()

# Aquí s'ha deduït la potència de la columna com a objecte, mentre que s'hauria d'haver deduït com a float64.
# Això pot ser perquè algunes de les files contenen valors no numèrics a la columna de potència.
# Una opció per fer-ho és forçar la conversió de la columna de cavalls de potència en numèrica,
# que hauria de convertir els valors no numèrics en valors nuls.
```

```python
autos['horsepower'] = pd.to_numeric(autos['horsepower'], errors='coerce')
autos.info()
```

```python
autos[autos.horsepower.isnull()]
```

```python
autos = autos.dropna(subset = ['horsepower'])
autos[autos.horsepower.isnull()]
```
